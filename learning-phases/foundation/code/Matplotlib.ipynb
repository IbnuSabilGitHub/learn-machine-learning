# Matplotlib

Matplotlib adalah library python untuk visualisasi data. pada ml ini digunakan untuk `Mengeklorasi Data`, `Melihat distribusi dan korelasi fitur`, dan `Menampilkan hasil prediksi model` 

## Import Matplotlib

```python
import matplotlib.pyplot as plt
```

## 1. Line Plot

```python
import numpy as np

# sample data random
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.plot(x) # plot garis
plt.title("Line Graph")
plt.xlabel("X-axis") # label sumbu x
plt.grid() # tambahkan grid
plt.show() # tampilkan plot
```

## 2. Scatter Plot

```python
plt.scatter(x,y, color="red") # plot scatter berwarna merah dengan x dan y
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y")
plt.show()
```

## 3. Histogram

## 4. Visualisasi DataFrame dari Pandas

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "Nama": ["Adit", "Denis", "Upin", "Ipin"],
    "Nilai": [90, 5, 75, 75]
})

plt.bar(df["Nama"], df["Nilai"], color="orange") # plot bar dengan nama dan nilai
plt.title("Nilai Siswa")
plt.ylabel("Nilai")
plt.show()
```

## 5. Visualisasi untuk Machine Learning


```python
# Regression
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
y_pred = 2 * X

plt.scatter(X, y, label="Data Asli")
plt.plot(X, y_pred, color="green", label="Prediksi")
plt.legend()
plt.title("Linear Regression")
plt.show()
```
